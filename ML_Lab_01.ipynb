```python
def mcp_and(x1, x2):
    w1, w2 = 1, 1
    theta = 2
    weighted_sum = w1 * x1 + w2 * x2
    output = 1 if weighted_sum >= theta else 0
    return output
for x1 in [0, 1]:
    for x2 in [0, 1]:
        print(f"AND({x1}, {x2}) = {mcp_and(x1, x2)}")
```

```python
def mcp_or(x1, x2):
    w1, w2 = 1, 1
    theta = 1
    weighted_sum = w1 * x1 + w2 * x2
    output = 1 if weighted_sum >= theta else 0
    return output
for x1 in [0, 1]:
    for x2 in [0, 1]:
        print(f"OR({x1}, {x2}) = {mcp_or(x1, x2)}")
```

```python
def mcp_xor(x1, x2):
    S = x1 + x2
    if S >= 1 and S < 2:
        return 1
    else:
        return 0

for x1 in [0, 1]:
    for x2 in [0, 1]:
        print(f"XOR({x1}, {x2}) = {mcp_xor(x1, x2)}")
```
